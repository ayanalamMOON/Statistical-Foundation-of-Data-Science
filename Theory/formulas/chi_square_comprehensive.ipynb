## Advanced Mathematical Derivations and Worked Examples

### Enhanced Mathematical Treatment of Chi-Square Tests

This section provides comprehensive mathematical derivations with step-by-step calculations and detailed worked examples for each type of chi-square test.

### Comprehensive Worked Example: Test of Independence

**Problem**: A pharmaceutical company wants to test if there's an association between age group and response to a new medication.

**Data**: Clinical trial with 300 patients

| Age Group     | Improved | No Change | Worsened | Total |
|---------------|----------|-----------|----------|-------|
| Young (18-35) | 45       | 25        | 10       | 80    |
| Middle (36-55)| 60       | 35        | 15       | 110   |
| Senior (56+)  | 55       | 40        | 15       | 110   |
| **Total**     | 160      | 100       | 40       | 300   |

#### Step-by-Step Mathematical Solution

**Step 1: Set up hypotheses**
- $H_0$: Age group and medication response are independent
- $H_1$: Age group and medication response are associated

Mathematically, independence means:
$$P(\text{Response}_j | \text{Age}_i) = P(\text{Response}_j) \text{ for all } i,j$$

**Step 2: Calculate expected frequencies under independence**

The expected frequency for cell $(i,j)$ under independence is:
$$E_{ij} = \frac{(\text{Row}_i \text{ Total}) \times (\text{Column}_j \text{ Total})}{\text{Grand Total}}$$

**Detailed calculations**:

For Young-Improved: $E_{11} = \frac{80 \times 160}{300} = \frac{12800}{300} = 42.67$

For Young-No Change: $E_{12} = \frac{80 \times 100}{300} = \frac{8000}{300} = 26.67$

For Young-Worsened: $E_{13} = \frac{80 \times 40}{300} = \frac{3200}{300} = 10.67$

For Middle-Improved: $E_{21} = \frac{110 \times 160}{300} = \frac{17600}{300} = 58.67$

For Middle-No Change: $E_{22} = \frac{110 \times 100}{300} = \frac{11000}{300} = 36.67$

For Middle-Worsened: $E_{23} = \frac{110 \times 40}{300} = \frac{4400}{300} = 14.67$

For Senior-Improved: $E_{31} = \frac{110 \times 160}{300} = \frac{17600}{300} = 58.67$

For Senior-No Change: $E_{32} = \frac{110 \times 100}{300} = \frac{11000}{300} = 36.67$

For Senior-Worsened: $E_{33} = \frac{110 \times 40}{300} = \frac{4400}{300} = 14.67$

**Step 3: Expected frequencies table**

| Age Group     | Improved | No Change | Worsened | Total |
|---------------|----------|-----------|----------|-------|
| Young (18-35) | 42.67    | 26.67     | 10.67    | 80    |
| Middle (36-55)| 58.67    | 36.67     | 14.67    | 110   |
| Senior (56+)  | 58.67    | 36.67     | 14.67    | 110   |
| **Total**     | 160.00   | 100.00    | 40.00    | 300   |

**Step 4: Calculate chi-square components**

For each cell $(i,j)$: $\frac{(O_{ij} - E_{ij})^2}{E_{ij}}$

| Cell | $O_{ij}$ | $E_{ij}$ | $O_{ij} - E_{ij}$ | $(O_{ij} - E_{ij})^2$ | $\frac{(O_{ij} - E_{ij})^2}{E_{ij}}$ |
|------|----------|----------|-------------------|----------------------|--------------------------------------|
| (1,1)| 45       | 42.67    | 2.33              | 5.43                 | 0.127                                |
| (1,2)| 25       | 26.67    | -1.67             | 2.79                 | 0.105                                |
| (1,3)| 10       | 10.67    | -0.67             | 0.45                 | 0.042                                |
| (2,1)| 60       | 58.67    | 1.33              | 1.77                 | 0.030                                |
| (2,2)| 35       | 36.67    | -1.67             | 2.79                 | 0.076                                |
| (2,3)| 15       | 14.67    | 0.33              | 0.11                 | 0.007                                |
| (3,1)| 55       | 58.67    | -3.67             | 13.47                | 0.230                                |
| (3,2)| 40       | 36.67    | 3.33              | 11.09                | 0.302                                |
| (3,3)| 15       | 14.67    | 0.33              | 0.11                 | 0.007                                |

**Step 5: Calculate test statistic**

$$\chi^2 = \sum_{i=1}^{3} \sum_{j=1}^{3} \frac{(O_{ij} - E_{ij})^2}{E_{ij}}$$

$$\chi^2 = 0.127 + 0.105 + 0.042 + 0.030 + 0.076 + 0.007 + 0.230 + 0.302 + 0.007 = 0.926$$

**Step 6: Degrees of freedom**

$$df = (r-1)(c-1) = (3-1)(3-1) = 2 \times 2 = 4$$

**Step 7: Critical value and p-value**

At $\alpha = 0.05$: $\chi^2_{0.05,4} = 9.488$

Since $\chi^2 = 0.926 < 9.488$, we fail to reject $H_0$.

**Step 8: Conclusion**

There is insufficient evidence to conclude that age group and medication response are associated. The observed differences could reasonably be due to random variation.

**Step 9: Effect size calculation**

Cramér's V: $V = \sqrt{\frac{\chi^2}{n \times \min(r-1, c-1)}} = \sqrt{\frac{0.926}{300 \times 2}} = \sqrt{0.00154} = 0.039$

This indicates a negligible association even if it were statistically significant.

### Advanced Mathematical Analysis: Power and Effect Size

#### Statistical Power in Chi-Square Tests

**Definition**: Power is the probability of correctly rejecting a false null hypothesis.

$$\text{Power} = P(\text{Reject } H_0 | H_0 \text{ is false}) = 1 - \beta$$

#### Power Calculation for Goodness of Fit Test

**Mathematical Framework**:

Let the true probabilities be $p_1^*, p_2^*, \ldots, p_k^*$ and the hypothesized probabilities be $p_1^0, p_2^0, \ldots, p_k^0$.

The **non-centrality parameter** is:
$$\lambda = n \sum_{i=1}^k \frac{(p_i^* - p_i^0)^2}{p_i^0}$$

Under the alternative hypothesis, the test statistic follows a **non-central chi-square distribution**:
$$\chi^2 \sim \chi^2(df, \lambda)$$

**Power calculation**:
$$\text{Power} = P(\chi^2(df, \lambda) > \chi^2_{\alpha, df})$$

#### Worked Example: Power Analysis

**Problem**: Calculate the power to detect a deviation from uniform distribution where one category has probability 0.3 instead of 0.2 (with 4 other categories at 0.175 each), using $n = 100$ and $\alpha = 0.05$.

**Solution**:

**Step 1: Define hypotheses**
- $H_0$: $p_1 = p_2 = p_3 = p_4 = p_5 = 0.2$ (uniform)
- $H_1$: $p_1 = 0.3, p_2 = p_3 = p_4 = p_5 = 0.175$

**Step 2: Calculate non-centrality parameter**
$$\lambda = 100 \sum_{i=1}^5 \frac{(p_i^* - p_i^0)^2}{p_i^0}$$

$$= 100 \left[ \frac{(0.3 - 0.2)^2}{0.2} + 4 \times \frac{(0.175 - 0.2)^2}{0.2} \right]$$

$$= 100 \left[ \frac{0.01}{0.2} + 4 \times \frac{0.000625}{0.2} \right]$$

$$= 100 \left[ 0.05 + 4 \times 0.003125 \right] = 100 \times 0.0625 = 6.25$$

**Step 3: Find critical value**
$df = k - 1 = 4$, $\chi^2_{0.05,4} = 9.488$

**Step 4: Calculate power**
Using the non-central chi-square distribution with $\lambda = 6.25$:
$$\text{Power} = P(\chi^2(4, 6.25) > 9.488) \approx 0.52$$

This means we have a 52% chance of detecting this particular departure from uniformity.

#### Effect Size Measures

#### 1. Cramér's V

For contingency tables:
$$V = \sqrt{\frac{\chi^2}{n \times \min(r-1, c-1)}}$$

**Interpretation**:
- $V = 0$: No association
- $V = 1$: Perfect association
- Cohen's conventions: 0.1 (small), 0.3 (medium), 0.5 (large)

#### 2. Phi Coefficient (for 2×2 tables)

$$\phi = \sqrt{\frac{\chi^2}{n}} = \frac{ad - bc}{\sqrt{(a+b)(c+d)(a+c)(b+d)}}$$

For a 2×2 table:
$$\begin{pmatrix} a & b \\ c & d \end{pmatrix}$$

#### 3. Contingency Coefficient

$$C = \sqrt{\frac{\chi^2}{\chi^2 + n}}$$

Note: $C$ has an upper bound less than 1, making it less interpretable than Cramér's V.

#### Worked Example: Effect Size Calculation

**Given**: 2×3 contingency table with $\chi^2 = 8.42$, $n = 200$

**Calculate all effect size measures**:

**Cramér's V**:
$$V = \sqrt{\frac{8.42}{200 \times \min(2-1, 3-1)}} = \sqrt{\frac{8.42}{200 \times 1}} = \sqrt{0.0421} = 0.205$$

**Contingency Coefficient**:
$$C = \sqrt{\frac{8.42}{8.42 + 200}} = \sqrt{\frac{8.42}{208.42}} = \sqrt{0.0404} = 0.201$$

**Interpretation**: Small to medium effect size indicating a weak to moderate association.

#### Sample Size Determination

**Formula for minimum sample size** (goodness of fit test):

$$n = \frac{(\chi^2_{\alpha,df} + \chi^2_{\beta,df})^2}{\sum_{i=1}^k \frac{(p_i^* - p_i^0)^2}{p_i^0}}$$

Where:
- $\chi^2_{\alpha,df}$ is the critical value at significance level $\alpha$
- $\chi^2_{\beta,df}$ is the critical value corresponding to power $1-\beta$

#### Worked Example: Sample Size Calculation

**Problem**: How many observations are needed to detect a deviation from uniform distribution (5 categories) where one category has probability 0.25 instead of 0.2, with power 0.8 and $\alpha = 0.05$?

**Solution**:

**Step 1: Identify parameters**
- $k = 5$, $df = 4$
- $\alpha = 0.05$, $\chi^2_{0.05,4} = 9.488$
- Power $= 0.8$, so $\beta = 0.2$, $\chi^2_{0.2,4} = 1.649$

**Step 2: Calculate effect size**
$$\sum_{i=1}^5 \frac{(p_i^* - p_i^0)^2}{p_i^0} = \frac{(0.25-0.2)^2}{0.2} + 4 \times \frac{(0.1875-0.2)^2}{0.2}$$

$$= \frac{0.0025}{0.2} + 4 \times \frac{0.000156}{0.2} = 0.0125 + 0.00312 = 0.01562$$

**Step 3: Calculate required sample size**
$$n = \frac{(9.488 + 1.649)^2}{0.01562} = \frac{(11.137)^2}{0.01562} = \frac{124.03}{0.01562} \approx 794$$

Therefore, approximately 794 observations are needed to achieve 80% power.

### Advanced Chi-Square Applications and Extensions

#### 1. McNemar's Test (Paired Chi-Square)

**Purpose**: Test for marginal homogeneity in paired categorical data.

**Mathematical Framework**:

For a 2×2 table of paired observations:

$$\begin{pmatrix}
a & b \\
c & d
\end{pmatrix}$$

Where $a, d$ are concordant pairs and $b, c$ are discordant pairs.

**Test Statistic**:
$$\chi^2 = \frac{(b-c)^2}{b+c}$$

**With continuity correction**:
$$\chi^2 = \frac{(|b-c|-1)^2}{b+c}$$

**Worked Example**:

A study examines preference changes after treatment:

|                | Post-Treatment |       |
|----------------|----------------|-------|
| **Pre-Treatment** | Prefer A | Prefer B |
| Prefer A       | 25       | 15       |
| Prefer B       | 10       | 30       |

**Calculation**:
- $b = 15$ (changed from A to B)
- $c = 10$ (changed from B to A)

$$\chi^2 = \frac{(15-10)^2}{15+10} = \frac{25}{25} = 1.0$$

With $df = 1$ and $\alpha = 0.05$, $\chi^2_{0.05,1} = 3.841$.

Since $1.0 < 3.841$, no significant change in preferences.

#### 2. Chi-Square Test for Trend (Cochran-Armitage Test)

**Purpose**: Test for linear trend in proportions across ordered categories.

**Mathematical Setup**:

For $k$ ordered groups with sample sizes $n_1, n_2, \ldots, n_k$ and success counts $r_1, r_2, \ldots, r_k$:

**Test Statistic**:
$$\chi^2_{\text{trend}} = \frac{[n\sum s_i r_i - (\sum s_i)(\sum r_i)]^2}{(\sum r_i)(\sum n_i - \sum r_i)[\sum n_i \sum s_i^2 - (\sum s_i)^2]}$$

Where $s_i$ are scores assigned to the ordered categories and $n = \sum n_i$.

**Worked Example**:

Test for dose-response relationship:

| Dose (mg) | Score ($s_i$) | Responders ($r_i$) | Total ($n_i$) | Proportion |
|-----------|---------------|-------------------|---------------|------------|
| 0         | 0             | 8                 | 50            | 0.16       |
| 10        | 1             | 15                | 50            | 0.30       |
| 20        | 2             | 22                | 50            | 0.44       |
| 30        | 3             | 28                | 50            | 0.56       |

**Calculations**:
- $n = 200$, $\sum r_i = 73$, $\sum s_i = 6$
- $\sum s_i r_i = 0(8) + 1(15) + 2(22) + 3(28) = 143$
- $\sum s_i^2 = 0 + 1 + 4 + 9 = 14$

$$\chi^2_{\text{trend}} = \frac{[200(143) - 6(73)]^2}{73(127)[200(14) - 36]} = \frac{[28600 - 438]^2}{73(127)(2764)} = 36.77$$

With $df = 1$, this is highly significant ($p < 0.001$), indicating a strong dose-response relationship.

#### 3. Mantel-Haenszel Test

**Purpose**: Test for association while controlling for confounding variables.

**Mathematical Framework**:

For $k$ 2×2 contingency tables (strata):

$$\begin{pmatrix}
a_i & b_i \\
c_i & d_i
\end{pmatrix} \quad i = 1, 2, \ldots, k$$

**Test Statistic**:
$$\chi^2_{MH} = \frac{\left(\sum_{i=1}^k a_i - \sum_{i=1}^k E[a_i]\right)^2}{\sum_{i=1}^k \text{Var}[a_i]}$$

Where:
- $E[a_i] = \frac{(a_i + b_i)(a_i + c_i)}{n_i}$
- $\text{Var}[a_i] = \frac{(a_i + b_i)(c_i + d_i)(a_i + c_i)(b_i + d_i)}{n_i^2(n_i - 1)}$

#### 4. Pearson Residuals and Standardized Residuals

**Purpose**: Identify which cells contribute most to chi-square statistic.

**Pearson Residuals**:
$$r_{ij} = \frac{O_{ij} - E_{ij}}{\sqrt{E_{ij}}}$$

**Standardized Residuals**:
$$z_{ij} = \frac{O_{ij} - E_{ij}}{\sqrt{E_{ij}(1-p_{i\cdot})(1-p_{\cdot j})}}$$

**Adjusted Standardized Residuals**:
$$d_{ij} = \frac{O_{ij} - E_{ij}}{\sqrt{E_{ij}(1-p_{i\cdot})(1-p_{\cdot j})}}$$

**Interpretation**:
- $|d_{ij}| > 2$: Significant at $\alpha = 0.05$
- $|d_{ij}| > 2.58$: Significant at $\alpha = 0.01$

#### 5. Exact Tests for Small Samples

**Fisher's Exact Test** (for 2×2 tables):

$$P = \frac{\binom{a+b}{a}\binom{c+d}{c}}{\binom{n}{a+c}}$$

**Freeman-Halton Test** (extension to larger tables):

$$P = \frac{n! \prod_{i,j} n_{ij}!}{\prod_i n_{i\cdot}! \prod_j n_{\cdot j}! \prod_{i,j} n_{ij}!}$$

#### 6. Log-Linear Models

**Mathematical Framework**:

For a contingency table, model the expected frequencies as:

$$\log E_{ij} = \mu + \lambda_i^A + \lambda_j^B + \lambda_{ij}^{AB}$$

**Independence Model**: $\lambda_{ij}^{AB} = 0$ for all $i,j$

**Likelihood Ratio Chi-Square**:
$$G^2 = 2\sum_{i,j} O_{ij} \log\left(\frac{O_{ij}}{E_{ij}}\right)$$

This provides an alternative to Pearson's chi-square with similar asymptotic properties.

#### 7. Goodness of Fit for Continuous Distributions

**Procedure**:
1. Divide the range into $k$ intervals
2. Calculate expected frequencies using the CDF
3. Apply standard chi-square test

**Example**: Testing normality with $\mu = 50$, $\sigma = 10$, $n = 100$

| Interval | Observed | $P(\text{interval})$ | Expected | $(O-E)^2/E$ |
|----------|----------|---------------------|----------|-------------|
| $< 40$   | 12       | 0.1587              | 15.87    | 0.944       |
| $40-50$  | 35       | 0.3413              | 34.13    | 0.022       |
| $50-60$  | 38       | 0.3413              | 34.13    | 0.439       |
| $> 60$   | 15       | 0.1587              | 15.87    | 0.048       |

$$\chi^2 = 0.944 + 0.022 + 0.439 + 0.048 = 1.453$$

With $df = 4 - 1 - 2 = 1$ (estimated $\mu$ and $\sigma$), this suggests good fit to normal distribution.

### Computational Methods and Implementation

#### 1. Algorithmic Calculation of Chi-Square Statistics

**Step-by-Step Algorithm**:

```
Input: Contingency table O[i,j] with r rows and c columns

Step 1: Calculate marginal totals
   row_totals[i] = Σⱼ O[i,j]
   col_totals[j] = Σᵢ O[i,j]
   grand_total = ΣᵢΣⱼ O[i,j]

Step 2: Calculate expected frequencies
   For i = 1 to r:
       For j = 1 to c:
           E[i,j] = (row_totals[i] × col_totals[j]) / grand_total

Step 3: Compute chi-square statistic
   chi_square = 0
   For i = 1 to r:
       For j = 1 to c:
           chi_square += (O[i,j] - E[i,j])² / E[i,j]

Step 4: Calculate degrees of freedom
   df = (r - 1) × (c - 1)

Step 5: Determine p-value
   p_value = P(χ²(df) ≥ chi_square)
```

#### 2. Numerical Computation of Chi-Square Distribution

**Probability Density Function**:
$$f(x; k) = \frac{1}{2^{k/2}\Gamma(k/2)} x^{k/2-1} e^{-x/2}$$

**Cumulative Distribution Function** (using incomplete gamma function):
$$F(x; k) = \frac{\gamma(k/2, x/2)}{\Gamma(k/2)}$$

Where $\gamma(s,x) = \int_0^x t^{s-1} e^{-t} dt$ is the lower incomplete gamma function.

**Series Expansion for CDF**:
$$F(x; k) = \sum_{i=0}^{\infty} \frac{(-1)^i (x/2)^{k/2+i}}{i! \Gamma(k/2+i+1)}$$

**Worked Numerical Example**:

Calculate $P(\chi^2_3 \leq 7.815)$:

Using the relationship with Poisson distribution:
$$P(\chi^2_{2n} \leq x) = e^{-x/2} \sum_{k=0}^{n-1} \frac{(x/2)^k}{k!}$$

For $k=3$ (odd df), we use:
$$P(\chi^2_3 \leq 7.815) = P(Z \leq \sqrt{7.815}) - P(Z \leq -\sqrt{7.815}) + \text{correction terms}$$

Approximation using Wilson-Hilferty transformation:
$$h = \frac{2}{9k}, \quad z = \frac{(x/k)^{1/3} - (1-h)}{\sqrt{h}}$$

For $x = 7.815$, $k = 3$:
- $h = \frac{2}{27} = 0.0741$
- $z = \frac{(7.815/3)^{1/3} - 0.9259}{0.272} = 0.995$

Therefore, $P(\chi^2_3 \leq 7.815) \approx \Phi(0.995) = 0.84$.

#### 3. Bootstrap Methods for Chi-Square Tests

**Parametric Bootstrap**:

```
For goodness of fit test with null hypothesis H₀: data ~ F₀

1. Calculate observed chi-square statistic χ²ₒᵦₛ
2. For b = 1 to B (bootstrap samples):
   a. Generate sample of size n from F₀
   b. Calculate chi-square statistic χ²ᵦ
3. Bootstrap p-value = (1/B) Σ I(χ²ᵦ ≥ χ²ₒᵦₛ)
```

**Non-parametric Bootstrap for Independence Test**:

```
For independence test of variables X and Y

1. Calculate observed chi-square statistic χ²ₒᵦₛ
2. For b = 1 to B:
   a. Permute the Y values randomly
   b. Form new contingency table
   c. Calculate chi-square statistic χ²ᵦ
3. Bootstrap p-value = (1/B) Σ I(χ²ᵦ ≥ χ²ₒᵦₛ)
```

#### 4. Computational Complexity and Optimization

**Time Complexity**:
- Basic chi-square calculation: O(rc) where r×c is table size
- Expected frequency calculation: O(rc)
- p-value computation: O(1) using pre-computed tables or O(log n) using series

**Memory Complexity**: O(rc) for storing the contingency table

**Optimization Strategies**:

1. **Sparse Matrix Representation** (for large, sparse tables):
   ```
   Store only non-zero entries as (row, col, value) triplets
   Memory: O(nnz) instead of O(rc)
   ```

2. **Vectorized Computation**:
   ```
   chi_square = sum((observed - expected)² / expected)
   Use element-wise operations instead of loops
   ```

3. **Numerical Stability**:
   ```
   For small expected frequencies, use:
   - Yates' continuity correction
   - Williams' correction factor: q = 1 + (r-1)(c-1)/(6n)
   - Adjusted statistic: χ²/q
   ```

#### 5. Monte Carlo Methods

**Monte Carlo p-value estimation**:

```
Procedure: MC_ChiSquare_Test(observed_table, iterations)
   chi_sq_obs = calculate_chi_square(observed_table)
   count_extreme = 0
   
   For i = 1 to iterations:
       simulated_table = simulate_under_H0(marginal_totals)
       chi_sq_sim = calculate_chi_square(simulated_table)
       if chi_sq_sim ≥ chi_sq_obs:
           count_extreme += 1
   
   p_value = count_extreme / iterations
   return p_value
```

**Confidence Interval for Monte Carlo p-value**:

Standard error: $SE = \sqrt{\frac{p(1-p)}{B}}$

95% CI: $p \pm 1.96 \sqrt{\frac{p(1-p)}{B}}$

For $B = 10,000$ and $p = 0.05$:
$SE = \sqrt{\frac{0.05 \times 0.95}{10,000}} = 0.0022$

#### 7. Handling Edge Cases

**Zero Frequencies**:
- Problem: Division by zero in $(O-E)^2/E$
- Solution: Add small constant $\delta$ or use exact methods

**Small Expected Frequencies**:
- Rule: If any $E_{ij} < 5$, chi-square approximation may be poor
- Solutions:
  1. Combine categories
  2. Use Fisher's exact test
  3. Use likelihood ratio test

**Large Contingency Tables**:
- Problem: Multiple testing issue
- Solution: Bonferroni correction or FDR control

**Computational Precision**:
- Use double precision arithmetic
- Check for numerical overflow in large samples
- Implement log-space calculations when necessary

## 1. Introduction to Chi-Square {#introduction}

The **Chi-square test** is a statistical hypothesis test used to determine whether there is a significant association between categorical variables or whether observed frequencies differ significantly from expected frequencies.

### Key Applications:
- **Goodness of Fit**: Testing if sample data fits a particular distribution
- **Test of Independence**: Testing if two categorical variables are independent
- **Test of Homogeneity**: Testing if different populations have the same distribution

### Assumptions:
1. Data must be in frequency counts (not percentages)
2. Categories must be mutually exclusive
3. Expected frequency in each cell should be at least 5
4. Observations must be independent

## 2. Mathematical Foundation {#mathematical-foundation}

### Chi-Square Statistic Formula

The fundamental chi-square statistic is defined as:

$$\chi^2 = \sum_{i=1}^{k} \frac{(O_i - E_i)^2}{E_i}$$

Where:
- $\chi^2$ = Chi-square test statistic
- $O_i$ = Observed frequency in category $i$
- $E_i$ = Expected frequency in category $i$
- $k$ = Number of categories

### Mathematical Intuition Behind the Formula

**Why do we square the differences?**
1. **Sign Independence**: $(O_i - E_i)^2$ ensures that positive and negative deviations contribute equally
2. **Penalty for Large Deviations**: Squaring amplifies larger deviations more than smaller ones
3. **Statistical Properties**: The sum of squared standardized normal variables follows a chi-square distribution

**Why do we divide by $E_i$?**
1. **Standardization**: Larger expected frequencies naturally have larger absolute deviations
2. **Relative Importance**: A deviation of 5 from an expected 10 is more significant than a deviation of 5 from an expected 100
3. **Variance Stabilization**: For Poisson-distributed counts, the variance equals the mean, so dividing by $E_i$ standardizes the variance

### Step-by-Step Derivation of the Chi-Square Statistic

**Step 1: Starting with the Multinomial Distribution**

For $n$ independent trials with $k$ possible outcomes, the probability of observing frequencies $(O_1, O_2, \ldots, O_k)$ is:

$$P(O_1, O_2, \ldots, O_k) = \frac{n!}{O_1! O_2! \cdots O_k!} p_1^{O_1} p_2^{O_2} \cdots p_k^{O_k}$$

**Step 2: Large Sample Approximation**

For large $n$, each $O_i$ is approximately normally distributed:
- $E[O_i] = np_i$
- $Var[O_i] = np_i(1-p_i) \approx np_i$ when $p_i$ is small

**Step 3: Standardization**

The standardized deviation is:
$$Z_i = \frac{O_i - E_i}{\sqrt{E_i}} \approx N(0,1)$$

**Step 4: Chi-Square Formation**

Since $Z_i^2 \sim \chi^2(1)$ for standard normal $Z_i$:
$$\chi^2 = \sum_{i=1}^{k} Z_i^2 = \sum_{i=1}^{k} \frac{(O_i - E_i)^2}{E_i}$$

### Worked Example: Understanding the Components

**Example**: Consider a simple 3-category case with $n = 30$ observations

| Category | Observed ($O_i$) | Expected ($E_i$) | $O_i - E_i$ | $(O_i - E_i)^2$ | $\frac{(O_i - E_i)^2}{E_i}$ |
|----------|------------------|------------------|-------------|-----------------|------------------------------|
| A        | 12               | 10               | 2           | 4               | 0.4                          |
| B        | 8                | 10               | -2          | 4               | 0.4                          |
| C        | 10               | 10               | 0           | 0               | 0.0                          |
| **Total**| 30               | 30               | 0           | 8               | **0.8**                      |

**Mathematical Calculation**:
$$\chi^2 = \frac{(12-10)^2}{10} + \frac{(8-10)^2}{10} + \frac{(10-10)^2}{10} = \frac{4}{10} + \frac{4}{10} + \frac{0}{10} = 0.8$$

### Degrees of Freedom: Mathematical Justification

The degrees of freedom represent the number of **independent** pieces of information available for estimating parameters.

#### For Goodness of Fit Test: $df = k - 1 - p$

**Derivation**:
1. We have $k$ observed frequencies: $O_1, O_2, \ldots, O_k$
2. **Constraint 1**: $\sum_{i=1}^k O_i = n$ (total sample size is fixed)
3. **Constraint 2**: If we estimate $p$ parameters from the data, we lose $p$ degrees of freedom
4. **Independent observations**: $k - 1 - p$

**Example**: Testing if data follows a normal distribution
- $k = 6$ categories (intervals)
- $p = 2$ parameters estimated (mean and variance)
- $df = 6 - 1 - 2 = 3$

#### For Independence Test: $df = (r-1)(c-1)$

**Derivation**:
1. We have an $r \times c$ contingency table with $rc$ cells
2. **Row constraints**: $\sum_{j=1}^c O_{ij} = n_{i\cdot}$ for $i = 1, 2, \ldots, r$ (gives $r$ constraints)
3. **Column constraints**: $\sum_{i=1}^r O_{ij} = n_{\cdot j}$ for $j = 1, 2, \ldots, c$ (gives $c$ constraints)
4. **Grand total constraint**: $\sum_{i=1}^r \sum_{j=1}^c O_{ij} = n$ (1 constraint, but already counted)
5. **Independent cells**: $rc - (r-1) - (c-1) - 1 = (r-1)(c-1)$

### Alternative Perspective: Linear Algebra Approach

The chi-square statistic can be written in matrix form:

$$\chi^2 = (\mathbf{O} - \mathbf{E})^T \mathbf{W} (\mathbf{O} - \mathbf{E})$$

Where:
- $\mathbf{O} = (O_1, O_2, \ldots, O_k)^T$ is the observed frequency vector
- $\mathbf{E} = (E_1, E_2, \ldots, E_k)^T$ is the expected frequency vector
- $\mathbf{W} = \text{diag}(1/E_1, 1/E_2, \ldots, 1/E_k)$ is the weight matrix

This formulation highlights that chi-square is a **weighted sum of squared deviations**.

## 3. Chi-Square Distribution Derivation {#distribution-derivation}

### Theoretical Foundation

If $Z_1, Z_2, \ldots, Z_n$ are independent standard normal random variables:

$$Z_i \sim N(0,1)$$

Then the sum of their squares follows a chi-square distribution:

$$\chi^2 = Z_1^2 + Z_2^2 + \cdots + Z_n^2 \sim \chi^2(n)$$

### Complete Mathematical Derivation of Chi-Square Distribution

#### Step 1: Starting with Standard Normal Variables

Let $Z \sim N(0,1)$ with probability density function:
$$f_Z(z) = \frac{1}{\sqrt{2\pi}} e^{-z^2/2}$$

#### Step 2: Transformation to Chi-Square (Single Variable)

For $Y = Z^2$, we need to find the PDF of $Y$. Using the transformation method:

If $y = z^2$, then $z = \pm\sqrt{y}$ and $\frac{dz}{dy} = \pm\frac{1}{2\sqrt{y}}$

The PDF of $Y = Z^2$ is:
$$f_Y(y) = f_Z(\sqrt{y}) \cdot \frac{1}{2\sqrt{y}} + f_Z(-\sqrt{y}) \cdot \frac{1}{2\sqrt{y}}$$

Since $f_Z(\sqrt{y}) = f_Z(-\sqrt{y})$ (symmetry of normal distribution):
$$f_Y(y) = 2 \cdot \frac{1}{\sqrt{2\pi}} e^{-y/2} \cdot \frac{1}{2\sqrt{y}} = \frac{1}{\sqrt{2\pi y}} e^{-y/2}$$

This can be rewritten as:
$$f_Y(y) = \frac{1}{2^{1/2} \Gamma(1/2)} y^{1/2-1} e^{-y/2}$$

This is the chi-square distribution with $\nu = 1$ degree of freedom.

#### Step 3: Moment Generating Function Approach

For $Z \sim N(0,1)$, the MGF of $Y = Z^2$ is:
$$M_Y(t) = E[e^{tZ^2}] = \int_{-\infty}^{\infty} e^{tz^2} \frac{1}{\sqrt{2\pi}} e^{-z^2/2} dz$$

$$= \frac{1}{\sqrt{2\pi}} \int_{-\infty}^{\infty} e^{-z^2(1/2-t)} dz$$

For $t < 1/2$, this integral converges to:
$$M_Y(t) = (1-2t)^{-1/2}$$

#### Step 4: Sum of Independent Chi-Square Variables

If $Y_1, Y_2, \ldots, Y_n$ are independent $\chi^2(1)$ variables, then:
$$\sum_{i=1}^n Y_i \sim \chi^2(n)$$

**Proof using MGFs**:
$$M_{\sum Y_i}(t) = \prod_{i=1}^n M_{Y_i}(t) = \prod_{i=1}^n (1-2t)^{-1/2} = (1-2t)^{-n/2}$$

This is the MGF of $\chi^2(n)$.

### Probability Density Function

The general PDF of chi-square distribution with $\nu$ degrees of freedom is:

$$f(x; \nu) = \frac{1}{2^{\nu/2} \Gamma(\nu/2)} x^{\nu/2-1} e^{-x/2}, \quad x \geq 0$$

#### Derivation of the PDF

**Step 1: Gamma Function Connection**

The chi-square distribution is a special case of the Gamma distribution:
$$\chi^2(\nu) = \text{Gamma}\left(\frac{\nu}{2}, \frac{1}{2}\right)$$

**Step 2: Gamma Distribution PDF**

For $X \sim \text{Gamma}(\alpha, \beta)$:
$$f(x) = \frac{\beta^\alpha}{\Gamma(\alpha)} x^{\alpha-1} e^{-\beta x}$$

**Step 3: Substitution**

With $\alpha = \nu/2$ and $\beta = 1/2$:
$$f(x) = \frac{(1/2)^{\nu/2}}{\Gamma(\nu/2)} x^{\nu/2-1} e^{-x/2} = \frac{1}{2^{\nu/2} \Gamma(\nu/2)} x^{\nu/2-1} e^{-x/2}$$

### Properties with Detailed Derivations

#### Mean: $E[\chi^2] = \nu$

**Derivation**:
Using the definition $\chi^2 = \sum_{i=1}^\nu Z_i^2$ where $Z_i \sim N(0,1)$:

$$E[\chi^2] = E\left[\sum_{i=1}^\nu Z_i^2\right] = \sum_{i=1}^\nu E[Z_i^2]$$

For $Z \sim N(0,1)$: $E[Z^2] = \text{Var}(Z) + (E[Z])^2 = 1 + 0^2 = 1$

Therefore: $E[\chi^2] = \sum_{i=1}^\nu 1 = \nu$

#### Variance: $\text{Var}[\chi^2] = 2\nu$

**Derivation**:
$$\text{Var}[\chi^2] = \text{Var}\left[\sum_{i=1}^\nu Z_i^2\right] = \sum_{i=1}^\nu \text{Var}[Z_i^2]$$

For $Z \sim N(0,1)$: 
- $E[Z^4] = 3$ (fourth moment of standard normal)
- $\text{Var}[Z^2] = E[Z^4] - (E[Z^2])^2 = 3 - 1^2 = 2$

Therefore: $\text{Var}[\chi^2] = \sum_{i=1}^\nu 2 = 2\nu$

#### Moment Generating Function: $M(t) = (1-2t)^{-\nu/2}$ for $t < \frac{1}{2}$

**Derivation**:
$$M(t) = E[e^{t\chi^2}] = E\left[e^{t\sum_{i=1}^\nu Z_i^2}\right] = \prod_{i=1}^\nu E[e^{tZ_i^2}]$$

Since each $Z_i^2 \sim \chi^2(1)$ with MGF $(1-2t)^{-1/2}$:
$$M(t) = \prod_{i=1}^\nu (1-2t)^{-1/2} = (1-2t)^{-\nu/2}$$

### Worked Example: Chi-Square with 3 Degrees of Freedom

**Given**: $\chi^2 \sim \chi^2(3)$

**PDF**: 
$$f(x) = \frac{1}{2^{3/2} \Gamma(3/2)} x^{3/2-1} e^{-x/2} = \frac{1}{2\sqrt{2} \cdot \frac{\sqrt{\pi}}{2}} x^{1/2} e^{-x/2}$$

Since $\Gamma(3/2) = \frac{1}{2}\Gamma(1/2) = \frac{\sqrt{\pi}}{2}$:
$$f(x) = \frac{1}{\sqrt{2\pi}} \sqrt{x} e^{-x/2}$$

**Mean**: $E[\chi^2] = 3$

**Variance**: $\text{Var}[\chi^2] = 2 \times 3 = 6$

**Standard Deviation**: $\sigma = \sqrt{6} \approx 2.45$

### Relationship to Other Distributions

#### Connection to Normal Distribution
$$\lim_{\nu \to \infty} \frac{\chi^2(\nu) - \nu}{\sqrt{2\nu}} \sim N(0,1)$$

#### Connection to F-Distribution
If $X \sim \chi^2(\nu_1)$ and $Y \sim \chi^2(\nu_2)$ are independent:
$$F = \frac{X/\nu_1}{Y/\nu_2} \sim F(\nu_1, \nu_2)$$

#### Connection to t-Distribution
If $Z \sim N(0,1)$ and $V \sim \chi^2(\nu)$ are independent:
$$t = \frac{Z}{\sqrt{V/\nu}} \sim t(\nu)$$

## 4. Goodness of Fit Test {#goodness-of-fit}

### Purpose
Test whether a sample comes from a population with a specific distribution.

### Hypotheses
- $H_0$: The sample follows the specified distribution
- $H_1$: The sample does not follow the specified distribution

### Expected Frequencies Calculation

For a discrete distribution with probabilities $p_1, p_2, \ldots, p_k$:

$$E_i = n \cdot p_i$$

where $n$ is the total sample size.

### Mathematical Derivation

Under $H_0$, each observed frequency $O_i$ follows a multinomial distribution:

$$\frac{O_i - E_i}{\sqrt{E_i}} \approx N(0,1)$$

Therefore:

$$\chi^2 = \sum_{i=1}^{k} \left(\frac{O_i - E_i}{\sqrt{E_i}}\right)^2 = \sum_{i=1}^{k} \frac{(O_i - E_i)^2}{E_i}$$

follows a chi-square distribution with $df = k-1$ degrees of freedom.

### Detailed Mathematical Derivation

#### Step 1: Multinomial Distribution Foundation

For $n$ independent trials with $k$ possible outcomes having probabilities $p_1, p_2, \ldots, p_k$, the joint probability mass function is:

$$P(O_1 = o_1, O_2 = o_2, \ldots, O_k = o_k) = \frac{n!}{o_1! o_2! \cdots o_k!} p_1^{o_1} p_2^{o_2} \cdots p_k^{o_k}$$

where $\sum_{i=1}^k o_i = n$ and $\sum_{i=1}^k p_i = 1$.

#### Step 2: Asymptotic Properties

For large $n$, each $O_i$ is approximately normal:
- $E[O_i] = np_i = E_i$
- $\text{Var}[O_i] = np_i(1-p_i) \approx np_i = E_i$ (when $p_i$ is small)
- $\text{Cov}[O_i, O_j] = -np_ip_j$ for $i \neq j$

#### Step 3: Standardization and Quadratic Form

Define the standardized deviations:
$$Z_i = \frac{O_i - E_i}{\sqrt{E_i}}$$

The chi-square statistic becomes:
$$\chi^2 = \sum_{i=1}^k Z_i^2$$

#### Step 4: Constraint and Degrees of Freedom

The constraint $\sum_{i=1}^k O_i = n$ implies:
$$\sum_{i=1}^k (O_i - E_i) = n - n = 0$$

This creates a linear dependence among the $Z_i$'s, reducing the effective degrees of freedom to $k-1$.

### Worked Example 1: Testing for Uniform Distribution

**Problem**: A spinner with 5 equal sections is spun 100 times. Test if the spinner is fair.

**Observed Data**: Section 1: 25, Section 2: 18, Section 3: 22, Section 4: 20, Section 5: 15

**Step-by-Step Solution**:

**Step 1: Set up hypotheses**
- $H_0$: The spinner is fair ($p_1 = p_2 = p_3 = p_4 = p_5 = 0.2$)
- $H_1$: The spinner is not fair

**Step 2: Calculate expected frequencies**
$$E_i = n \cdot p_i = 100 \times 0.2 = 20 \text{ for all } i$$

**Step 3: Calculate chi-square components**

| Section | $O_i$ | $E_i$ | $O_i - E_i$ | $(O_i - E_i)^2$ | $\frac{(O_i - E_i)^2}{E_i}$ |
|---------|-------|-------|-------------|-----------------|------------------------------|
| 1       | 25    | 20    | 5           | 25              | 1.25                         |
| 2       | 18    | 20    | -2          | 4               | 0.20                         |
| 3       | 22    | 20    | 2           | 4               | 0.20                         |
| 4       | 20    | 20    | 0           | 0               | 0.00                         |
| 5       | 15    | 20    | -5          | 25              | 1.25                         |

**Step 4: Calculate test statistic**
$$\chi^2 = 1.25 + 0.20 + 0.20 + 0.00 + 1.25 = 2.90$$

**Step 5: Determine degrees of freedom**
$$df = k - 1 = 5 - 1 = 4$$

**Step 6: Find critical value and p-value**
At $\alpha = 0.05$: $\chi^2_{0.05,4} = 9.488$

**Step 7: Make decision**
Since $\chi^2 = 2.90 < 9.488$, we fail to reject $H_0$. The spinner appears to be fair.

### Worked Example 2: Testing for Poisson Distribution

**Problem**: Test if the number of accidents per day follows a Poisson distribution.

**Data**: Days with 0 accidents: 15, 1 accident: 25, 2 accidents: 12, 3 accidents: 6, 4+ accidents: 2 (Total: 60 days)

**Step-by-Step Solution**:

**Step 1: Estimate Poisson parameter**
Total accidents = $0(15) + 1(25) + 2(12) + 3(6) + 4(2) = 0 + 25 + 24 + 18 + 8 = 75$
Estimated rate: $\hat{\lambda} = \frac{75}{60} = 1.25$

**Step 2: Calculate expected frequencies**
For Poisson($\lambda = 1.25$):
- $P(X = 0) = e^{-1.25} \times \frac{1.25^0}{0!} = 0.2865$
- $P(X = 1) = e^{-1.25} \times \frac{1.25^1}{1!} = 0.3581$
- $P(X = 2) = e^{-1.25} \times \frac{1.25^2}{2!} = 0.2238$
- $P(X = 3) = e^{-1.25} \times \frac{1.25^3}{3!} = 0.0933$
- $P(X \geq 4) = 1 - P(X \leq 3) = 0.0383$

Expected frequencies: $E_i = 60 \times P(X = i)$

**Step 3: Calculation table**

| Accidents | $O_i$ | $P(X = i)$ | $E_i = 60P(X = i)$ | $(O_i - E_i)^2$ | $\frac{(O_i - E_i)^2}{E_i}$ |
|-----------|-------|------------|-------------------|-----------------|------------------------------|
| 0         | 15    | 0.2865     | 17.19             | 4.80            | 0.279                        |
| 1         | 25    | 0.3581     | 21.49             | 12.32           | 0.573                        |
| 2         | 12    | 0.2238     | 13.43             | 2.04            | 0.152                        |
| 3         | 6     | 0.0933     | 5.60              | 0.16            | 0.029                        |
| 4+        | 2     | 0.0383     | 2.30              | 0.09            | 0.039                        |

**Step 4: Calculate test statistic**
$$\chi^2 = 0.279 + 0.573 + 0.152 + 0.029 + 0.039 = 1.072$$

**Step 5: Degrees of freedom**
$$df = k - 1 - p = 5 - 1 - 1 = 3$$
(We estimated 1 parameter: $\lambda$)

**Step 6: Critical value**
At $\alpha = 0.05$: $\chi^2_{0.05,3} = 7.815$

**Step 7: Conclusion**
Since $\chi^2 = 1.072 < 7.815$, we fail to reject $H_0$. The data is consistent with a Poisson distribution.

## 5. Test of Independence {#test-of-independence}

### Purpose
Test whether two categorical variables are independent.

### Hypotheses
- $H_0$: Variables are independent (no association)
- $H_1$: Variables are not independent (there is association)

### Contingency Table

For a $r \times c$ contingency table:

| Variable B \ Variable A | $A_1$ | $A_2$ | $\cdots$ | $A_c$ | Total |
|-------------------------|-------|-------|----------|-------|---------|
| $B_1$                   | $n_{11}$ | $n_{12}$ | $\cdots$ | $n_{1c}$ | $n_{1\cdot}$ |
| $B_2$                   | $n_{21}$ | $n_{22}$ | $\cdots$ | $n_{2c}$ | $n_{2\cdot}$ |
| $\vdots$                | $\vdots$ | $\vdots$ | $\ddots$ | $\vdots$ | $\vdots$ |
| $B_r$                   | $n_{r1}$ | $n_{r2}$ | $\cdots$ | $n_{rc}$ | $n_{r\cdot}$ |
| **Total**               | $n_{\cdot 1}$ | $n_{\cdot 2}$ | $\cdots$ | $n_{\cdot c}$ | $n$ |

### Expected Frequencies Under Independence

If variables are independent:

$$P(A_j \cap B_i) = P(A_j) \cdot P(B_i)$$

The expected frequency for cell $(i,j)$ is:

$$E_{ij} = \frac{n_{i\cdot} \cdot n_{\cdot j}}{n}$$

### Chi-Square Statistic

$$\chi^2 = \sum_{i=1}^{r} \sum_{j=1}^{c} \frac{(O_{ij} - E_{ij})^2}{E_{ij}}$$

with $df = (r-1)(c-1)$ degrees of freedom.

## 6. Test of Homogeneity {#test-of-homogeneity}

### Purpose
Test whether different populations have the same distribution for a categorical variable.

### Hypotheses
- $H_0$: All populations have the same distribution
- $H_1$: At least one population has a different distribution

### Mathematical Framework

For $k$ populations and $c$ categories:

$$H_0: p_{1j} = p_{2j} = \cdots = p_{kj} = p_j \text{ for all } j = 1, 2, \ldots, c$$

The test statistic is identical to the independence test:

$$\chi^2 = \sum_{i=1}^{k} \sum_{j=1}^{c} \frac{(O_{ij} - E_{ij})^2}{E_{ij}}$$

with $df = (k-1)(c-1)$ degrees of freedom.

## 7. Step-by-Step Calculations {#step-by-step}

### Example 1: Goodness of Fit Test

**Problem**: A die is rolled 60 times. Test if the die is fair.

**Observed frequencies**: Face 1: 8, Face 2: 12, Face 3: 10, Face 4: 15, Face 5: 7, Face 6: 8

````python
# Step 1: Define the observed frequencies
observed = np.array([8, 12, 10, 15, 7, 8])
n = sum(observed)

print(f"Observed frequencies: {observed}")
print(f"Total observations: {n}")

# Step 2: Calculate expected frequencies (fair die assumption)
# For a fair die, each face should appear with probability 1/6
k = len(observed)  # number of categories
expected_prob = 1/k  # probability for each face
expected = n * expected_prob

print(f"\nExpected probability for each face: {expected_prob:.4f}")
print(f"Expected frequency for each face: {expected}")

# Create a DataFrame for better visualization
df_goodness = pd.DataFrame({
    'Face': [f'Face {i}' for i in range(1, k+1)],
    'Observed (O_i)': observed,
    'Expected (E_i)': [expected] * k,
    'O_i - E_i': observed - expected,
    '(O_i - E_i)^2': (observed - expected)**2,
    '(O_i - E_i)^2 / E_i': (observed - expected)**2 / expected
})

print("\nStep-by-step calculation table:")
print(df_goodness)
````

````python
# Step 3: Calculate chi-square statistic manually
chi_square_manual = sum((observed - expected)**2 / expected)

print(f"\nManual Chi-square calculation:")
print(f"χ² = Σ[(O_i - E_i)² / E_i]")
print(f"χ² = {' + '.join([f'{val:.4f}' for val in (observed - expected)**2 / expected])}")
print(f"χ² = {chi_square_manual:.4f}")

# Step 4: Determine degrees of freedom
df = k - 1  # k categories - 1
print(f"\nDegrees of freedom: df = k - 1 = {k} - 1 = {df}")

# Step 5: Find critical value and p-value
alpha = 0.05
critical_value = chi2.ppf(1 - alpha, df)
p_value = 1 - chi2.cdf(chi_square_manual, df)

print(f"\nAt α = {alpha}:")
print(f"Critical value: χ²_{df,{alpha}} = {critical_value:.4f}")
print(f"P-value: {p_value:.4f}")

# Step 6: Make decision
print(f"\nDecision:")
if chi_square_manual > critical_value:
    print(f"χ² = {chi_square_manual:.4f} > {critical_value:.4f} = χ²_critical")
    print("Reject H₀: The die is not fair")
else:
    print(f"χ² = {chi_square_manual:.4f} ≤ {critical_value:.4f} = χ²_critical")
    print("Fail to reject H₀: The die appears to be fair")

if p_value < alpha:
    print(f"P-value = {p_value:.4f} < {alpha} = α")
    print("Reject H₀ (consistent with critical value approach)")
else:
    print(f"P-value = {p_value:.4f} ≥ {alpha} = α")
    print("Fail to reject H₀ (consistent with critical value approach)")
````

````python
# Verify with scipy.stats
chi2_scipy, p_scipy = stats.chisquare(observed)

print(f"\nVerification with scipy.stats.chisquare():")
print(f"χ² statistic: {chi2_scipy:.4f}")
print(f"P-value: {p_scipy:.4f}")
print(f"\nManual calculation matches scipy: {abs(chi_square_manual - chi2_scipy) < 1e-10}")
````

### Example 2: Test of Independence

**Problem**: Test if there's an association between gender and preferred beverage.

**Data**: Survey of 200 people

````python
# Step 1: Create contingency table
contingency_table = np.array([
    [25, 35, 15],  # Male: Coffee, Tea, Juice
    [30, 40, 55]   # Female: Coffee, Tea, Juice
])

row_labels = ['Male', 'Female']
col_labels = ['Coffee', 'Tea', 'Juice']

# Create DataFrame for better presentation
df_contingency = pd.DataFrame(contingency_table,
                             index=row_labels,
                             columns=col_labels)

print("Observed Contingency Table:")
print(df_contingency)

# Calculate marginal totals
row_totals = contingency_table.sum(axis=1)
col_totals = contingency_table.sum(axis=0)
grand_total = contingency_table.sum()

print(f"\nRow totals: {row_totals}")
print(f"Column totals: {col_totals}")
print(f"Grand total: {grand_total}")

# Add totals to the DataFrame
df_with_totals = df_contingency.copy()
df_with_totals['Total'] = row_totals
df_with_totals.loc['Total'] = list(col_totals) + [grand_total]

print("\nContingency Table with Marginal Totals:")
print(df_with_totals)
````

````python
# Step 2: Calculate expected frequencies under independence
# Formula: E_ij = (row_total_i × col_total_j) / grand_total

r, c = contingency_table.shape
expected_table = np.zeros((r, c))

print("Expected Frequencies Calculation:")
print("Formula: E_ij = (Row_i_Total × Col_j_Total) / Grand_Total\n")

for i in range(r):
    for j in range(c):
        expected_table[i, j] = (row_totals[i] * col_totals[j]) / grand_total
        print(f"E_{i+1}{j+1} = ({row_totals[i]} × {col_totals[j]}) / {grand_total} = {expected_table[i, j]:.2f}")

# Create DataFrame for expected frequencies
df_expected = pd.DataFrame(expected_table,
                          index=row_labels,
                          columns=col_labels)

print("\nExpected Frequencies Table:")
print(df_expected.round(2))
````

````python
# Step 3: Calculate chi-square statistic manually
print("Chi-square Calculation:")
print("Formula: χ² = Σ Σ [(O_ij - E_ij)² / E_ij]\n")

chi_square_components = []
chi_square_manual_indep = 0

# Create detailed calculation table
calculation_data = []

for i in range(r):
    for j in range(c):
        observed_val = contingency_table[i, j]
        expected_val = expected_table[i, j]
        difference = observed_val - expected_val
        squared_diff = difference**2
        contribution = squared_diff / expected_val

        chi_square_manual_indep += contribution
        chi_square_components.append(contribution)

        calculation_data.append({
            'Cell': f"{row_labels[i]}-{col_labels[j]}",
            'Observed (O_ij)': observed_val,
            'Expected (E_ij)': round(expected_val, 2),
            'O_ij - E_ij': round(difference, 2),
            '(O_ij - E_ij)²': round(squared_diff, 2),
            '(O_ij - E_ij)²/E_ij': round(contribution, 4)
        })

        print(f"Cell ({row_labels[i]}, {col_labels[j]}): ({observed_val} - {expected_val:.2f})² / {expected_val:.2f} = {contribution:.4f}")

# Create calculation DataFrame
df_calculation = pd.DataFrame(calculation_data)
print("\nDetailed Calculation Table:")
print(df_calculation)

print(f"\nχ² = {' + '.join([f'{comp:.4f}' for comp in chi_square_components])}")
print(f"χ² = {chi_square_manual_indep:.4f}")
````

````python
# Step 4: Degrees of freedom and statistical inference
df_indep = (r - 1) * (c - 1)
print(f"\nDegrees of freedom: df = (r-1)(c-1) = ({r}-1)({c}-1) = {df_indep}")

# Critical value and p-value
alpha = 0.05
critical_value_indep = chi2.ppf(1 - alpha, df_indep)
p_value_indep = 1 - chi2.cdf(chi_square_manual_indep, df_indep)

print(f"\nAt α = {alpha}:")
print(f"Critical value: χ²_{df_indep,{alpha}} = {critical_value_indep:.4f}")
print(f"P-value: {p_value_indep:.4f}")

# Decision
print(f"\nDecision:")
if chi_square_manual_indep > critical_value_indep:
    print(f"χ² = {chi_square_manual_indep:.4f} > {critical_value_indep:.4f} = χ²_critical")
    print("Reject H₀: Gender and beverage preference are associated")
else:
    print(f"χ² = {chi_square_manual_indep:.4f} ≤ {critical_value_indep:.4f} = χ²_critical")
    print("Fail to reject H₀: Gender and beverage preference are independent")

# Effect size - Cramér's V
cramers_v = np.sqrt(chi_square_manual_indep / (grand_total * min(r-1, c-1)))
print(f"\nEffect Size (Cramér's V): {cramers_v:.4f}")

# Interpretation of effect size
if cramers_v < 0.1:
    effect_interpretation = "negligible"
elif cramers_v < 0.3:
    effect_interpretation = "small"
elif cramers_v < 0.5:
    effect_interpretation = "medium"
else:
    effect_interpretation = "large"

print(f"Effect size interpretation: {effect_interpretation} association")
````

````python
# Verify with scipy.stats
chi2_scipy_indep, p_scipy_indep, dof_scipy, expected_scipy = stats.chi2_contingency(contingency_table)

print(f"\nVerification with scipy.stats.chi2_contingency():")
print(f"χ² statistic: {chi2_scipy_indep:.4f}")
print(f"P-value: {p_scipy_indep:.4f}")
print(f"Degrees of freedom: {dof_scipy}")
print(f"\nExpected frequencies from scipy:")
print(pd.DataFrame(expected_scipy, index=row_labels, columns=col_labels).round(2))

print(f"\nManual calculation matches scipy: {abs(chi_square_manual_indep - chi2_scipy_indep) < 1e-10}")
````

## 8. Python Implementation {#python-implementation}

### Custom Functions for Chi-Square Tests

````python
def chi_square_goodness_of_fit(observed, expected=None, alpha=0.05):
    """
    Perform chi-square goodness of fit test with detailed output

    Parameters:
    observed (array-like): Observed frequencies
    expected (array-like): Expected frequencies (if None, assumes equal distribution)
    alpha (float): Significance level

    Returns:
    dict: Comprehensive test results
    """
    observed = np.array(observed)
    n = observed.sum()
    k = len(observed)

    if expected is None:
        expected = np.full(k, n/k)  # Equal distribution
    else:
        expected = np.array(expected)

    # Check assumptions
    min_expected = expected.min()
    assumptions_met = min_expected >= 5

    # Calculate chi-square statistic
    chi_square = np.sum((observed - expected)**2 / expected)

    # Degrees of freedom
    df = k - 1

    # Critical value and p-value
    critical_value = chi2.ppf(1 - alpha, df)
    p_value = 1 - chi2.cdf(chi_square, df)

    # Decision
    reject_null = chi_square > critical_value

    # Create results dictionary
    results = {
        'chi_square': chi_square,
        'degrees_of_freedom': df,
        'p_value': p_value,
        'critical_value': critical_value,
        'alpha': alpha,
        'reject_null': reject_null,
        'observed': observed,
        'expected': expected,
        'residuals': observed - expected,
        'standardized_residuals': (observed - expected) / np.sqrt(expected),
        'assumptions_met': assumptions_met,
        'min_expected_frequency': min_expected
    }

    return results

def print_chi_square_results(results, test_type="Goodness of Fit"):
    """
    Print formatted chi-square test results
    """
    print(f"\n{'='*60}")
    print(f"CHI-SQUARE {test_type.upper()} TEST RESULTS")
    print(f"{'='*60}")

    print(f"\nTest Statistic: χ² = {results['chi_square']:.4f}")
    print(f"Degrees of Freedom: df = {results['degrees_of_freedom']}")
    print(f"P-value: {results['p_value']:.6f}")
    print(f"Critical Value (α = {results['alpha']}): {results['critical_value']:.4f}")

    print(f"\nHypotheses:")
    if test_type == "Goodness of Fit":
        print(f"H₀: Data follows the specified distribution")
        print(f"H₁: Data does not follow the specified distribution")
    elif test_type == "Independence":
        print(f"H₀: Variables are independent")
        print(f"H₁: Variables are not independent")

    print(f"\nDecision:")
    if results['reject_null']:
        print(f"✗ Reject H₀ (χ² = {results['chi_square']:.4f} > {results['critical_value']:.4f})")
        if test_type == "Goodness of Fit":
            print(f"  The data does not follow the specified distribution")
        elif test_type == "Independence":
            print(f"  There is significant association between variables")
    else:
        print(f"✓ Fail to reject H₀ (χ² = {results['chi_square']:.4f} ≤ {results['critical_value']:.4f})")
        if test_type == "Goodness of Fit":
            print(f"  The data is consistent with the specified distribution")
        elif test_type == "Independence":
            print(f"  Variables appear to be independent")

    print(f"\nAssumptions:")
    if results['assumptions_met']:
        print(f"✓ All expected frequencies ≥ 5 (minimum: {results['min_expected_frequency']:.2f})")
    else:
        print(f"⚠ Warning: Some expected frequencies < 5 (minimum: {results['min_expected_frequency']:.2f})")
        print(f"  Results may not be reliable")

    print(f"{'='*60}")

# Test the function with our die example
die_results = chi_square_goodness_of_fit([8, 12, 10, 15, 7, 8])
print_chi_square_results(die_results, "Goodness of Fit")
````

````python
def chi_square_independence(contingency_table, alpha=0.05):
    """
    Perform chi-square test of independence with detailed output

    Parameters:
    contingency_table (array-like): 2D array of observed frequencies
    alpha (float): Significance level

    Returns:
    dict: Comprehensive test results
    """
    observed = np.array(contingency_table)
    r, c = observed.shape

    # Calculate marginal totals
    row_totals = observed.sum(axis=1)
    col_totals = observed.sum(axis=0)
    grand_total = observed.sum()

    # Calculate expected frequencies
    expected = np.outer(row_totals, col_totals) / grand_total

    # Check assumptions
    min_expected = expected.min()
    assumptions_met = min_expected >= 5

    # Calculate chi-square statistic
    chi_square = np.sum((observed - expected)**2 / expected)

    # Degrees of freedom
    df = (r - 1) * (c - 1)

    # Critical value and p-value
    critical_value = chi2.ppf(1 - alpha, df)
    p_value = 1 - chi2.cdf(chi_square, df)

    # Decision
    reject_null = chi_square > critical_value

    # Effect size (Cramér's V)
    cramers_v = np.sqrt(chi_square / (grand_total * min(r-1, c-1)))

    # Standardized residuals
    standardized_residuals = (observed - expected) / np.sqrt(expected)

    # Create results dictionary
    results = {
        'chi_square': chi_square,
        'degrees_of_freedom': df,
        'p_value': p_value,
        'critical_value': critical_value,
        'alpha': alpha,
        'reject_null': reject_null,
        'observed': observed,
        'expected': expected,
        'row_totals': row_totals,
        'col_totals': col_totals,
        'grand_total': grand_total,
        'standardized_residuals': standardized_residuals,
        'cramers_v': cramers_v,
        'assumptions_met': assumptions_met,
        'min_expected_frequency': min_expected
    }

    return results

# Test the function with our beverage example
beverage_data = np.array([
    [25, 35, 15],  # Male
    [30, 40, 55]   # Female
])

independence_results = chi_square_independence(beverage_data)
print_chi_square_results(independence_results, "Independence")

print(f"\nEffect Size (Cramér's V): {independence_results['cramers_v']:.4f}")
if independence_results['cramers_v'] < 0.1:
    print("Effect size: Negligible association")
elif independence_results['cramers_v'] < 0.3:
    print("Effect size: Small association")
elif independence_results['cramers_v'] < 0.5:
    print("Effect size: Medium association")
else:
    print("Effect size: Large association")
````

## Summary and Key Formulas

### Essential Chi-Square Formulas

| Test Type | Formula | Degrees of Freedom | Application |
|-----------|---------|-------------------|-------------|
| **Goodness of Fit** | $\chi^2 = \sum_{i=1}^{k} \frac{(O_i - E_i)^2}{E_i}$ | $df = k - 1 - p$ | Test if data fits distribution |
| **Independence** | $\chi^2 = \sum_{i=1}^{r} \sum_{j=1}^{c} \frac{(O_{ij} - E_{ij})^2}{E_{ij}}$ | $df = (r-1)(c-1)$ | Test variable association |
| **Homogeneity** | $\chi^2 = \sum_{i=1}^{k} \sum_{j=1}^{c} \frac{(O_{ij} - E_{ij})^2}{E_{ij}}$ | $df = (k-1)(c-1)$ | Compare populations |

### Expected Frequency Calculations

- **Goodness of Fit**: $E_i = n \cdot p_i$
- **Independence/Homogeneity**: $E_{ij} = \frac{\text{Row}_i \times \text{Col}_j}{\text{Total}}$

### Effect Size

**Cramér's V**: $V = \sqrt{\frac{\chi^2}{n \cdot \min(r-1, c-1)}}$

- 0.1: Small effect
- 0.3: Medium effect  
- 0.5: Large effect

### Critical Assumptions

1. **Independence**: Observations must be independent
2. **Expected Frequencies**: Each cell should have $E_{ij} \geq 5$
3. **Sample Size**: Adequate sample size for reliable results
4. **Categorical Data**: Variables must be categorical (not continuous)

### Decision Rules

- **Reject $H_0$ if**: $\chi^2 > \chi^2_{\alpha, df}$ OR $p < \alpha$
- **Fail to reject $H_0$ if**: $\chi^2 \leq \chi^2_{\alpha, df}$ OR $p \geq \alpha$
